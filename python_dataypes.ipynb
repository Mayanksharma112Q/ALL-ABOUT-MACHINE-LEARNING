Basic Data Types in Python:
1. Integer
2. Floating point
3. Complex
4. Boolean
5. String

```python
# integers
a = 8
print(a)
type(a)
```

```python
# floating point
b = 2.3
print(b)
type(b)
```

```python
# complex numbers
c = 1 + 3j
print(c)
type(c)
```

Conversion of one data type to another

```python
# int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
# float to int
x = 5.88
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```

# Boolean

```python
a = 6<5
print(a)
type(a)
```

```python
b = 8>5
print(b)
type(b)
```

# Slicing

```python
my_string = "Programming"
print(my_string[1:5]) # Values from index 1 to 5-1 will be sliced
```

```python
# step means it will go 1 to 10-1 but every 2nd element or digit will be sliced
print(my_string[0:10:2])
```

# String concatenation

```python
word_1 = 'Machine'
word_2 =' Learning'

print(word_1 + word_2)
```
