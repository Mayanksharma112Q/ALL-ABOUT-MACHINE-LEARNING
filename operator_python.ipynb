1. Arithmetic operators

```python
num1 = 3
num2 = 5
# addition 
sum = num1+num2
print('sum =  ',sum)

# subtraction
sub = num1-num2
print('sub =  ',sub)

#Multiplication
product = num1*num2
print('product =  ',product)

#Division
Div = num1/num2
print('Quotient =  ',Div)

#Modulus
Mod = num1%num2
print('Remainder = ',Mod)

#Exponent
exp = num1**num2
print('Exponent = ',exp)
```

3. comparison

```python
a = 5
b = 10

print(a == b)   # equal to
print(a != b)   # not equal to
print(a > b)    # greater than
print(a < b)
print(a <= b)
print(a >= b)  
```

6. Membership operator

```python
a = 5
b = 10

c = [1,2,3,4,5]
print(a in c)
print(b in c)
```
