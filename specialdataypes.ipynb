# 1. List

```python
 #list should be included in the square brackets
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# lists can have multiple data types
my_list = [2,3,1.6,'english',True]
print(my_list)
```

### Lists are mutuable --> changeable

```python
# add elements to a list
my_list = [2,3,1.8,'hindi',True]
my_list.append(6)
print(my_list)
```

```python
# print elements of a list using their index
print(my_list[0])
print(my_list[3])
```

```python
# lists allow duplicate values
list_1 = [1,2,3,4,2,3,6,4,11,32]
print(list_1)
```

```python
print(len(list_1))
```

```python
# initiating an empty list
list_2 = []
print(list_2)
```

```python
list_2.append(5)
print(list_2)
```

```python
# dalete an item in a list
list_2 = [2,3,1.8,'english',True,6]
print(list_2)

del list_2[2]
print(list_2)
```

```python
# join two lists
list_3 = [1,2,3,4,5]
list_4 = [2,6,7,7,8]
list_5 = list_3+list_4
print(list_5)
```

### Tuple

```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
# tuple allows multiple data types
tuple_2 = (1,2,3.5,'Machine learning',False)
print(tuple_2)
```

```python
# converting a list to a tuple
my_list = [3,4,5,6]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[1])
```

# Tuples are immutable --> unchangeable

```python
my_tuple.append(6)
```

```python
print(len(my_tuple))
```

### Set

```python
# set --> Curly brackets
my_set = {1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
print(my_set[0])
```

```python
# convert a lsit to a set
list_5 = [4,5,6,7,8]

x = set(list_5)
print(x)
```

```python
# set does not allow duplicate values
set_3 = {1,2,3,4,5,1,2,3}
print(set_3)
```

## Dictionary

Key-value pair

```python
my_dictionary = {'name':'David','age':30,'country':'India'}
print(my_dictionary)
type(my_dictionary)
```

```python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['country'])
```

```python
# dictionary does not allow duplicate values
dictionary_2 = {'name':'David','age':30,'country':'India','name':'David','age':30,'country':'India'}
print(dictionary_2)
```
